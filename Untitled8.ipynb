```python
data1["Avg Monthly Long Distance Charges"]=data1["Avg Monthly Long Distance Charges"].fillna(0)
data1["Multiple Lines"]=data1["Multiple Lines"].fillna("No")
data1["Internet Type"]=data1["Internet Type"].fillna("No")
data1["Avg Monthly GB Download"]=data1["Avg Monthly GB Download"].fillna(0)
data1["Online Security"]=data1["Online Security"].fillna("No")
data1["Online Backup"]=data1["Online Backup"].fillna("No")
data1["Device Protection Plan"]=data1["Device Protection Plan"].fillna("No")
data1["Premium Tech Support"]=data1["Premium Tech Support"].fillna("No")
data1["Streaming TV"]=data1["Streaming TV"].fillna("No")
data1["Streaming Movies"]=data1["Streaming Movies"].fillna("No")
data1["Streaming Music"]=data1["Streaming Music"].fillna("No")
data1["Unlimited Data"]=data1["Unlimited Data"].fillna("No")
data1["Offer"]=data1["Offer"].fillna("No")
```

```python
data1
```

```python
data1["Customer Status"].value_counts()
```

```python
bins= [18,34,50,66,82]
labels = ['Young Adult','Adult','Mid-Age Adult','Senior Citizen']
data1['AgeGroup'] = pd.cut(data1['Age'], bins=bins, labels=labels, right=False)
```

```python
data1["AgeGroup"].value_counts()
```

```python
data1['Gender'] = data1['Gender'].map({'Male': 0, 'Female': 1})
data1['Customer Status'] = data1['Customer Status'].map({'Stayed': 0, 'Churned': 1})
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Display basic information about the dataset
print(data1.info())

# Show the first few rows of the dataset
print("\
First few rows of the dataset:")
print(data1.head())

# Display summary statistics
print("\
Summary statistics:")
print(data1.describe())

# Calculate churn rate
churn_rate = data1[data1['Customer Status'] == 'Churned'].shape[0] / data1.shape[0] * 100
print(f"\
Churn Rate: {churn_rate:.2f}%")
```

```python
# Visualize churn distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Customer Status', data=data1)
plt.title('Distribution of Customer Status')
plt.show()
```

```python
# Analyze the relationship between tenure and churn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Customer Status', y='Tenure in Months', data=data1)
plt.title('Tenure vs. Customer Status')
plt.show()
```

```python
# Analyze churn reasons
churn_reasons = data[data['Customer Status'] == 'Churned']['Churn Reason'].value_counts().head(10)
plt.figure(figsize=(12, 6))
churn_reasons.plot(kind='bar')
plt.title('Top 10 Churn Reasons')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Analyze the impact of contract type on churn
contract_churn = data1.groupby('Contract')['Customer Status'].value_counts(normalize=True).unstack()
contract_churn.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Customer Status by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Percentage')
plt.legend(title='Customer Status')
plt.tight_layout()
plt.show()

print("Analysis complete.")
```
